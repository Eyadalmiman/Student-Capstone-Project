Here we made a variable called "df" inside of it we put the .csv

```python
df = pd.read_csv("cleaned_data.csv")
```

Here we computed the mean and the standard devation of SalePrice, we firstly converted the dataframe into a numpy array then we did np.mean() and np.std() (Used Google search to find out how to convert the df into a numpy array)

```python
target = df["SalePrice"].to_numpy()
mean = np.mean(target)
std = np.std(target)

print(f"Mean of SalePrice : {mean}")
print(f"std of SalePrice : {std}")
```

Here we basically computed both Normal by hand broadcasting and the fast more efficent way which is StandardScaler and saw what was the difference of outputs between them (Refreshed my memory by looking at the course slides)

```python
dfS = df.copy()
col = dfS["SalePrice"] # We made the variable after a copied df so we can make sure that we dont mess up anything
col_SS = dfS["SalePrice"]

mean = col.mean()
sttd = col.std(ddof = 0) # This makes it match StandardScaler (this was a tip from gemini)
broadcasting = (col - mean) / sttd

std = StandardScaler()
dfS["SalePrice_SS"] = std.fit_transform(dfS[["SalePrice"]])

print(f"Normal broadcasting by hand : {broadcasting}")
print(f"StandardScaler : {dfS[["SalePrice_SS"]].agg(["mean", "std"])}")
```

Compute cosine similarity between the feature vectors of the highest-value and lowest-value
records (from course slides)

```python
def cosine_simularity(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))

feature = ["Gr Liv Area", "Garage Area", "Total Bsmt SF", "Overall Qual"]

high_val = df.loc[df["SalePrice"].idxmax(), feature].values
low_val = df.loc[df["SalePrice"].idxmin(), feature].values

simularity = cosine_simularity(high_val, low_val)

print(f"Cosine Simularity (High and Low) {simularity:.2f}")
```

Estimate a probability: e.g., what fraction of high-quality items have a target above a certain
threshold?

```python
High_qual = df["Garage Qual"] == "Gd"
high_qual_df = df[High_qual]

threshold = 240000

over_threshold = (high_qual_df["SalePrice"] > threshold).sum()

total_high_qual = len(high_qual_df)
probability = over_threshold / total_high_qual

print(f"Total of High Quality houses (Good consition and better) : {total_high_qual} houses.")
print(f"Number of houses above the threshold ({threshold}) : {over_threshold} houses")
print(f"Probability : {probability:.2f}")
```
